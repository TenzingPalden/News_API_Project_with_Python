```python
import requests
import pandas as pd
import csv
```

```python
#making a list with 47 numbers in it to use for the counter for the dataframe later. Could use an edited index but I want a seperate column.
nombre=[]
for x in range(1,48):
    nombre.append(x)
```

```python
#making a dataframe with a single column titled "Title".            
df=pd.DataFrame(columns=["Title"])
```

```python
#Requesting a specfic set of infromation from the News API with API KEY VARIABLE. Next converting the information into a dictionary
apikey="0c0bd82da34e4525971bd519eb64d133"
url=f"https://newsapi.org/v2/everything?q=blockchain&pageSize=47&apiKey={apikey}"
response=requests.get(url)
news=response.json()

#itterating through the dictionary to add the titles to the dataframe under the column "Titles"
for x in range(0, 47):
    df.loc[x]=news["articles"][x]["title"]
```

```python
#Viewing the dataframe to check that it looks okay. 
df
```

```python
number= len(df)
print(f"the len is : {number}")
```

```python
#adding additional columns. intigrating the list we made earlier into a column. Splitting the title into lists and counting them and then adding it onto the dataframe.
df["Position on chart Count"]=nombre
df["Word Count"]=df["Title"].str.split().str.len()
df["Character Count"]=df["Title"].str.len()
```

```python
#Viewing the dataframe to check that it looks okay. 
df
```

```python
#saves dataframe to a csv file without the index.
df.to_csv('G22Results.csv', index=False)
```

```python
#saving the average of the respective columns into their own variables.
AverageWord=df["Word Count"].mean()
AverageCharacter=df["Character Count"].mean()
```

```python
#printing the results of the averages of the columns into readble text for humans. Self explanitory cell.

print(f"The average for the word count is {AverageWord} words and the average for the character count is {AverageCharacter} characters.")
```
